```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Generate a sample signal
np.random.seed(0)
signal_length = 512
signal = np.cumsum(np.random.randn(signal_length))

# Define the wavelet and level of decomposition
wavelet = 'db6'  # Daubechies wavelet with one vanishing moment
level = 5

# Perform MODWT
coeffs = pywt.wavedec(signal, wavelet, level=level, mode='per')

# Plot the original signal and its MODWT coefficients
plt.figure(figsize=(10, 6))

# Plot the original signal
plt.subplot(level + 2, 1, 1)
plt.plot(signal, label='Original Signal')
plt.legend()

# Plot the MODWT coefficients
for i in range(level):
    plt.subplot(level + 2, 1, i + 2)
    plt.plot(coeffs[i], label=f'Level {i + 1}')
    plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
